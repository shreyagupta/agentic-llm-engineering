# Step 1

Prepare our data for fine-tuning in JSONL (JSON Lines) format and upload to OpenAI

````python
# First let's work on a good prompt for a Frontier model
# Notice that I'm removing the " to the nearest dollar"
# When we train our own models, we'll need to make the problem as easy as possible,
# but a Frontier model needs no such simplification.

def messages_for(item):
    system_message = "You estimate prices of items. Reply only with the price, no explanation"
    user_prompt = item.test_prompt().replace(" to the nearest dollar","").replace("\n\nPrice is $","")
    return [
        {"role": "system", "content": system_message},
        {"role": "user", "content": user_prompt},
        {"role": "assistant", "content": f"Price is ${item.price:.2f}"}
    ]
````

````python
# Convert the items into a list of json objects - a "jsonl" string
# Each row represents a message in the form:
# {"messages" : [{"role": "system", "content": "You estimate prices...


def make_jsonl(items):
    result = ""
    for item in items:
        messages = messages_for(item)
        messages_str = json.dumps(messages)
        result += '{"messages": ' + messages_str +'}\n'
    return result.strip()
````

````python
print(make_jsonl(train[:3]))
````

````python
# Convert the items into jsonl and write them to a file

def write_jsonl(items, filename):
    with open(filename, "w") as f:
        jsonl = make_jsonl(items)
        f.write(jsonl)
````

Excellent question ‚Äî the difference between `"rb"` and `"r"` when using `open()` is subtle but important.

Let‚Äôs break it down:

---

### üîç The difference

| Mode   | Meaning              | What you get when reading    | Typical use                                          |
| ------ | -------------------- | ---------------------------- | ---------------------------------------------------- |
| `"r"`  | **read text mode**   | `str` (decoded Unicode text) | Reading normal text files (JSON, CSV, etc.)          |
| `"rb"` | **read binary mode** | `bytes` (raw binary data)    | Reading binary files (images, compressed data, etc.) |

---

### üí° Example

**Using `"r"` (text mode):**

```python
with open("fine_tune_validation.jsonl", "r") as f:
    for line in f:
        print(type(line))   # <class 'str'>
```

* Each line you read is a **string** (`str`).
* Python automatically decodes bytes ‚Üí text using your system‚Äôs default encoding (usually UTF-8).

---

**Using `"rb"` (binary mode):**

```python
with open("fine_tune_validation.jsonl", "rb") as f:
    for line in f:
        print(type(line))   # <class 'bytes'>
```

* Each line is **raw bytes** (`b'...'`).
* No decoding is done ‚Äî you‚Äôd need to manually call `.decode("utf-8")` to get text:

  ```python
  text_line = line.decode("utf-8")
  ```

---

### üß† When to use each

‚úÖ Use `"r"` when:

* You‚Äôre reading **text-based files** like `.json`, `.jsonl`, `.csv`, `.txt`.
* You want strings for easy parsing with `json.loads(line)` or similar.

‚úÖ Use `"rb"` when:

* You‚Äôre reading **binary files** (images, audio, compressed archives).
* You‚Äôre uploading the file as binary (e.g., via API) where the raw bytes are needed.
* You want to ensure **no newline or encoding transformations** occur (some APIs expect binary streams).

---

### ‚öôÔ∏è In your specific case:

`fine_tune_validation.jsonl` is a **JSON Lines text file**, so you should almost always use:

```python
with open("fine_tune_validation.jsonl", "r", encoding="utf-8") as f:
```

Using `"rb"` is only necessary if the function you‚Äôre passing `f` to (e.g., an API uploader) explicitly expects **a binary stream**.

````python
# The prompt

def messages_for(item):
    system_message = "You estimate prices of items. Reply only with the price, no explanation"
    user_prompt = item.test_prompt().replace(" to the nearest dollar","").replace("\n\nPrice is $","")
    return [
        {"role": "system", "content": system_message},
        {"role": "user", "content": user_prompt},
        {"role": "assistant", "content": "Price is $"}
    ]
````

````python
# Try this out

messages_for(test[0])
````

````python
# A utility function to extract the price from a string

def get_price(s):
    s = s.replace('$','').replace(',','')
    match = re.search(r"[-+]?\d*\.\d+|\d+", s)
    return float(match.group()) if match else 0
````

````python
get_price("The price is roughly $99.99 because blah blah")
````

````python
# The function for gpt_fine_tuned is the same as that for gpt-4o-mini, the only difference being that we pass in our fine_tuned_model_name and not the gpt-4o-mini model name!

def gpt_fine_tuned(item):
    response = openai.chat.completions.create(
        model=fine_tuned_model_name,
        messages=messages_for(item),
        seed=42,
        max_tokens=7
    )
    reply = response.choices[0].message.content
    return get_price(reply)
````

````python
print(test[0].price)
print(gpt_fine_tuned(test[0]))
````

````python
print(test[0].test_prompt())
````

````python
Tester.test(gpt_fine_tuned, test)
````
