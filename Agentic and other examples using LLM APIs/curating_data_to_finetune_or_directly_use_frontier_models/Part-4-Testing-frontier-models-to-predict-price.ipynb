## First, the humble but mighty GPT-4o-mini

It's called mini, but it packs a punch. As seen later, its performance is quite at par with GPT_4o_frontier

```python
# First let's work on a good prompt for a Frontier model
# Notice that I'm removing the " to the nearest dollar"
# When we train our own models, we'll need to make the problem as easy as possible, 
# but a Frontier model needs no such simplification.

def messages_for(item):
    system_message = "You estimate prices of items. Reply only with the price, no explanation"
    user_prompt = item.test_prompt().replace(" to the nearest dollar","").replace("\n\nPrice is $","")
    return [
        {"role": "system", "content": system_message},
        {"role": "user", "content": user_prompt},
        {"role": "assistant", "content": "Price is $"}
    ]
```

```python
# let's check it for one sample point in test data

messages_for(test[0])
```

```python
# A utility function to extract the price from a string

def get_price(s):
    s = s.replace('$','').replace(',','')
    match = re.search(r"[-+]?\d*\.\d+|\d+", s)
    return float(match.group()) if match else 0
```

```python
#let's test it
get_price("The price is roughly $99.99 because blah blah")
```

```python
# The function for gpt-4o-mini

def gpt_4o_mini(item):
    response = openai.chat.completions.create(
        model="gpt-4o-mini", 
        messages=messages_for(item),
        seed=42,
        max_tokens=5
    )
    reply = response.choices[0].message.content
    return get_price(reply)
```

```python
def gpt_4o_frontier(item):
    response = openai.chat.completions.create(
        model="gpt-4o-2024-08-06", 
        messages=messages_for(item),
        seed=42,
        max_tokens=5
    )
    reply = response.choices[0].message.content
    return get_price(reply)
```

```python
#APIT was getting overloaded and function was failing, so here I demo how to overcome that

from anthropic._exceptions import OverloadedError
import time

def claude_4_point_5_sonnet(item,max_retries=10):
    
    messages = messages_for(item)
    system_message = messages[0]['content']
    messages = messages[1:]
    for attempt in range(max_retries):
        try:
            response = claude.messages.create(
                model="claude-sonnet-4-20250514",
                max_tokens=5,
                system=system_message,
                messages=messages
            )
            reply = response.content[0].text
            return get_price(reply)
        except OverloadedError:#anthropic.OverloadedError:
                if attempt < max_retries - 1:
                    wait_time = 2 ** attempt  # exponential backoff: 1s, 2s, 4s
                    print(f"API overloaded, retrying in {wait_time}s...")
                    time.sleep(wait_time)
                else:
                    print("API still overloaded after retries")
                    raise
```
